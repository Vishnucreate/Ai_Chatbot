```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords

# Download necessary NLTK resources
nltk.download('punkt')  # for tokenization
nltk.download('wordnet')  # for lemmatization
nltk.download('stopwords')  # for removing stopwords

# Function to read the content of the file
def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Load your text file
text = read_file('/content/100Kcharacters.txt')

# Tokenization
tokens = word_tokenize(text)
print("Tokens:", tokens[:10])  # Display the first 10 tokens

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token not in stop_words]
print("Tokens after removing stopwords:", filtered_tokens[:10])  # Display the first 10 filtered tokens

# Stemming
stemmer = PorterStemmer()
stems = [stemmer.stem(token) for token in filtered_tokens]
print("Stems:", stems[:10])  # Display the first 10 stems

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(token) for token in filtered_tokens]
print("Lemmas:", lemmas[:10])  # Display the first 10 lemmas
```

```python
# Step 1: Install the necessary library
!pip install transformers

# Step 2: Import the library
from transformers import pipeline
import os

# Step 3: Load the pre-trained model for question answering
model_name = 'bert-large-uncased-whole-word-masking-finetuned-squad'
qa_pipeline = pipeline('question-answering', model=model_name)

# Function to read the file content
def load_text(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        return file.read()

# Step 4: Load the content of your text file
file_path = '/content/100Kcharacters.txt'  # Adjust this path to your file's location
context = load_text(file_path)

# Step 5: Interactive question-answering loop
def interactive_qa():
    print("Enter 'quit' to exit.")
    while True:
        question = input("Ask a question: ")
        if question.lower() == 'quit':
            break
        response = qa_pipeline({'context': context, 'question': question})
        print("Answer:", response['answer'])

# Run the interactive QA system
interactive_qa()
```
